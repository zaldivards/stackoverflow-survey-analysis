# 1. What is the relation that has the parents' education on the education and salary of the respondents? 

## Data preparation

```python
# getting the variables that will be usefull to answer the question
variables = ['HighestEducationParents', 'Salary', 'EmploymentStatus', 
'CareerSatisfaction', 'JobSatisfaction', 'DiversityImportant', 'FormalEducation']
df = data[variables] 
```

```python
# checking for na values and the data types
df.info()
```

```python
# to do an unbiased analysis, all the variable will be removed based on the salary
df = df.dropna(subset=['Salary'])
df.info()
```

```python
df.isna().mean() * 100
```

You can see some of the categorical variables have missing values. As the missing percentages are relatively small, we'll fill it with the **mode** statistic.

```python
# filling the categorical variables
objects = df.select_dtypes('object')
objects = objects.fillna(objects.mode().squeeze())
objects.info()
```

The numerical variables JobSatisfaction and DiversityImportant have missing values. The missing percentage is insignificant,hence, a **mean** imputation would work very well.

```python
# filling the numerical variables
float_data = df.select_dtypes('float')
float_data = float_data.fillna(float_data.mean())
float_data.info()
```

```python
# joining both the numerical and categorical data 
df_ = pd.concat([objects, float_data], axis=1)
df_.info()
```

```python
parents = (df_['HighestEducationParents'].value_counts() / df_['HighestEducationParents'].count()).sort_values(ascending=False)
parents
```

```python
# PLOT-1

# measure with percentages the parents' education 
(df_['HighestEducationParents'].value_counts() / df_['HighestEducationParents'].count()).sort_values().plot.barh(title="Parents' education");
```

### Salaries based on the parents' education

```python
# PLOT-2

# Respondents' average salaries based on their parents' education 
df_.groupby('HighestEducationParents')['Salary'].mean().sort_values().plot.barh(title="Respondents' average salaries based on their parents' education");
```

### Salaries based on the repondents' education

```python
# PLOT-3

# Respondents' average salaries based on their education
df_.groupby('FormalEducation')['Salary'].mean().sort_values().plot.barh(title="Respondents' average salaries based on their education");
```

```python
# Relation between parents' and respondents' education
df_[['FormalEducation', 'HighestEducationParents']].apply(pd.Series.value_counts)
```

As we can see there are some mismatches in the responses, let's fix it to properly plot it

```python
# updating values so it match between the values of HighestEducationParents and FormalEducation
df_['HighestEducationParents'] = df_['HighestEducationParents'].replace('A doctoral degree', 'Doctoral degree')\
    .replace("A bachelor's degree", "Bachelor's degree")\
    .replace("A master's degree", "Master's degree")\
    .replace("A professional degree", "Professional degree")\
    .replace("High school", "Secondary school")\
    .replace("Some college/university study, no bachelor's degree", "Some college/university study without earning a bachelor's degree")\
    .replace("No education", "I never completed any formal education")
```

```python
# PLOT-4

# Comparison between respondents' and their parents' education
df_[['FormalEducation', 'HighestEducationParents']].apply(pd.Series.value_counts).sort_values('HighestEducationParents').plot.barh(title="Comparison between respondents' and their parents' education");
```

## Result evaluation

As you can see, the parents' education has a clear relation in the respondents' salaries. The respondents present a higher salary when their parents have higher education. However, this means nothing about the respondents' salaries based on their education. Take a look at the second element in **PLOT-3**, they have no professional degree, just elementary school studies. So interesting!!!

When we are talking about the highest salaries that a developer could seek, **a professional degree is often not necessary**.

Furthermore, we can conclude that the parents' education has an impact on the respondents' education. You can take a look at **PLOT-4** to make sure of it. As you can see, the number of respondents with higher degrees(except for the Doctoral degree) is bigger than the corresponding number of their parents.

# 2. Does overpayment increase the job satisfaction?

## Data preparation

```python
# variables that will be used to answer the current question
variables = ['Overpaid', 'JobSatisfaction', 'WorkPayCare', 'ProblemSolving', 'CompanyType', 'ProgramHobby']
df = data[variables]
```

```python
# understanding data types
df.info()
```

```python
# looking for null data
df.isna().mean() * 100
```

```python
# removing null values based on Overpaid column because it has too much null values(73%)
df.dropna(subset=['Overpaid'], inplace=True)
```

```python
# looking for null data
df.isna().mean() * 100
```

You can see the above categorical variables have almost **32%** of missing data. It is still acceptable, so we'll fill it with the **mode**

```python
def filler(x: pd.Series) -> str:
    """Fills the missing data of categorical variables

    Args:
        x: pandas Series data structure with elements of type <object> 
    Returns:
        The mode statistic of x
    """
    return x.fillna(x.mode()[0])
objects = ['WorkPayCare', 'ProblemSolving', 'CompanyType']
# fill categorical data with the mode
df[objects] = df[objects].apply(filler)
```

Just **2.7%** of the answers are null for job satisfaction variable. We'll fill those null answers with the **mean** of the existent ones.
This variable is numeric, the mean works well.

```python
# fill numerical data with the mean 
df['JobSatisfaction'] = df['JobSatisfaction'].fillna(df['JobSatisfaction'].mean())
```

```python
# looking for null values after removing and filling
df.isna().mean() * 100
```

```python
df.head()
```

```python
df.info()
```

```python
# Job satisfaction per Overpaid category
overpaid_df = df.groupby('Overpaid')['JobSatisfaction'].mean().sort_values(ascending=False)
overpaid_df
```

Definitely, the people that are overpaid have higher job satisfaction. However, the people that is somewhat underpaid doesn't seem unsatisfied with their job. Furthermore, is interesting to note that the people which is greatly overpaid have a job satisfaction less than the one of the people who are neither underpaid nor overpaid. Could the type of company or the interest in what they work on influence on their job satisfaction?

```python
# filtering by the category with interesting result
target = df.query("Overpaid == 'Greatly overpaid'")
```

```python
# Job satisfaction per type of company
companies_types = target.groupby('CompanyType')['JobSatisfaction'].mean().sort_values(ascending=False)
companies_types
```

```python
# Number of respondents with a certain type of company
target['CompanyType'].value_counts()
```

```python
# Job satisfaction per the interest in what they do and the salary 
workcare = target.groupby('WorkPayCare')['JobSatisfaction'].mean().sort_values(ascending=False)
workcare
```

```python
fig, ax = plt.subplots(2, 1)
plt.subplots_adjust(hspace=0.5)
ax[0].set_xlabel('Job satisfaction', fontsize=15)
ax[0].set_title('Job Satisfaction based on company type', fontsize=15)
ax[1].set_xlabel('Job satisfaction', fontsize=15)
ax[1].set_title('Job Satisfaction based on the interest on what the respondents work on', fontsize=15)

companies_types.plot.barh(ax=ax[0], fontsize=15)
workcare.plot.barh(ax=ax[1], fontsize=15);
```

## Result evaluation

First of all, remember the above aggregates are based only on the data of the respondents who are greatly overpaid.

There's and interest pattern. The companies that have job satisfaction between 6.1 and 6.7 are NGOs or government companies. However we cannot conclude nothing from it, there's no clear evidence and the number of company types is **skewed**. On the other side you can take a look at the answers related with the workPayCare. The people who answered **Strongly disagree** have a job satisfaction of 6.36, the lowest! This means that there's a proportion of respondents within this subset that are not entirely satisfied with their functions within their companies, no matter if they are overpaid.

So, the job satisfaction level of the people who are greatly overpaid **may be** influenced by the work environment. Moreover, the repondents' functions within their companies may have another factor of influence.

# 3. What is the percentage of people who are currently employed but looking for a new job? Is this people underpaid or their job satisfaction is low?

## Data preparation

```python
# using only the interesting variables
variables = ['EmploymentStatus', 'JobSeekingStatus', 'CompanyType', 'JobSatisfaction', 'Overpaid', 'HoursPerWeek']
df = data[variables]
```

```python
# checking dtypes
df.info()
```

```python
# Unique options and frequency of EmploymentStatus variable
df.EmploymentStatus.value_counts()
```

```python
# # Unique options and frequency of JobSeekingStatus variable
df.JobSeekingStatus.value_counts()
```

```python
# statuses of people who are employed in some way
employment_statuses = ['Employed full-time', 'Independent contractor, freelancer, or self-employed', 'Employed part-time']
```

```python
# filtering by the above statuses and people who are looking for a job
df_filtered = df[(df['EmploymentStatus'].isin(employment_statuses)) & (df['JobSeekingStatus'] == 'I am actively looking for a job')]
```

```python
# percentage of people who are employed and looking for a job
f'{(len(df_filtered) / len(df)) * 100:.2f}%'
```

From the entire dataset **6.46%** of respondents are employed but looking for a job. Let's check if these people are underpaid and how is their job satisfaction.

```python
# Percentages of people who are either underpaid or overpaid
overpayment = (df_filtered['Overpaid'].value_counts() / df_filtered['Overpaid'].count()) * 100
overpayment
```

```python
# PLOT-1

ax= overpayment.plot.barh()
ax.set_xlabel('Count', fontsize=15)
ax.set_ylabel('Overpayment answers', fontsize=15)
ax.set_title('Overpayment proportion of people who are employed but looking for a job', fontsize=15);
```

```python
# checking the proportion of null values
df_filtered.Overpaid.isna().mean() * 100
```

```python
satisfaction = df_filtered['JobSatisfaction'].value_counts().sort_index()
satisfaction
```

```python
# PLOT-2

ax = satisfaction.plot.bar(alpha=0.7, legend=True)
plt.axvline(df_filtered['JobSatisfaction'].mean(), color='r', label='Average satisfaction')
plt.legend(fontsize=15)
ax.set_xlabel('Job satisfaction', fontsize=15)
ax.set_ylabel('N. of people', fontsize=15)
ax.set_title('Job satisfaction rates', fontsize=15);
```

## Result evaluation

It is clear that most people in this subset are **underpaid**, **73%** aproximately(take a look at **PLOT-1**). Though this would be a good metric to conclude about the reasons, there is a lot of uncertainty because the variable "Overpaid" has **70%** of null values.  

As you can see in **PLOT-2**, the job satisfaction fractions are greater in the left side from the Average. This tell us that people within this subset often have lower job satisfaction than the average: **4.9**.
